

Building Model & Testing Accuracy



```python
target = df['quality']
features=df.drop('quality',axis=1)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
X_train, X_test, y_train, y_test = train_test_split(features,target, test_size = 0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
dt_predict = dt.predict(X_test)

dt_acc_score = accuracy_score(y_test, dt_predict)
print(dt_acc_score)
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=20)
random_forest.fit(X_train, y_train)

Y_prediction = random_forest.predict(X_test)

accuracy_randomforest = random_forest.score(X_train, y_train)
print((accuracy_randomforest))
```

We got an accuray of more than 99% using Random Forest Classifier and around 60% using Decision Tree Classifier.

```python
results = pd.DataFrame({"Actual": y_test, "Predicted": Y_prediction})
results
```

Now, We can very precisisely predict wine quality using Random Forest Classifier.
